```python
#8th AUGUST
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Used_Bikes.csv")
df.head(2)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
brands=["TVS","Yamaha","Bajaj","Hero","Royal Enfield","Honda","Suzuki","TVS","KTM","Harley-Davidson","Kawasaki","Mahindra","Hyosung","Benelli","Triumph","Ducati","BMW"]
df[df['brand'].isin(brands)]
```

OR

```python
#9th AUGUST
brand_dict={brand:i for i , brand in enumerate(list(df['brand'].value_counts().keys()),start=1)}
print(brand_dict)
df['brand']=df['brand'].map(brand_dict)     #do not execute it 2nd time
```

```python
df.head()
```

```python
df.drop(['bike_name','city'],axis=1,inplace=True)
df.head(2)
```

```python
x=df.drop(['price'],axis=1)
x
```

```python
y=df[['price']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(x,y,test_size=0.2)     #test_size=0.2 means test size is 20% and training size automatically becomes 80%
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression       #LinearRegression is a class of linear regression model
```

```python
lnr=LinearRegression()
```

```python
lnr.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
x_train
```

```python
lnr.score(x_train,y_train)*100      #achieved 73% accuracy
```

```python
lnr.score(x_test,y_test)*100        #unseen or test or validation data
```

OVERFIT MODEL => training score is high but testing score is low<br>
UNDERFIT MODEL => training score is low but testing score can be low or high

```python
#PREDICTION
prediction=lnr.predict(x_test)
```

```python
y_test['prediction']=prediction
y_test.head(20)
```
